```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
```

```python
trader=pd.read_csv("/content/historical_data.csv")
fg=pd.read_csv("/content/fear_greed_index.csv")
```

```python
# Trader Data
print(trader.shape)
print(trader.columns.tolist())
trader.info()
trader.describe()
trader.isnull().sum()
trader.head(10)

```

```python
# Fear & Greed
print(fg.shape)
print(fg.columns.tolist())
fg.info()
fg.head()
```

```python
# ── TRADER DATA ──────────────────────────────────────────
# Format is DD-MM-YYYY HH:MM  e.g. "18-03-2025 12:50"
trader['Timestamp IST'] = pd.to_datetime(trader['Timestamp IST'], format='%d-%m-%Y %H:%M')

# Extract date for merging
trader['date'] = pd.to_datetime(trader['Timestamp IST'].dt.date)

print("Trader date range:", trader['date'].min(), "to", trader['date'].max())
print("Sample dates:", trader['date'].head())

# ── FEAR & GREED DATA ─────────────────────────────────────
fg['date'] = pd.to_datetime(fg['date'])

print("\nFear/Greed date range:", fg['date'].min(), "to", fg['date'].max())
```

```python
merged = pd.merge(trader, fg, on='date', how='left')

print("Merged shape:", merged.shape)
print("Null sentiment count:", merged['classification'].isnull().sum())
merged[['Account', 'Coin', 'Closed PnL', 'Side', 'classification', 'date']].head()
```

```python
# Filter out 0 PnL rows (open/non-closed trades)
merged_filtered = merged[merged['Closed PnL'] != 0].copy()
print("Trades with actual PnL:", merged_filtered.shape[0])

order = ['Extreme Fear', 'Fear', 'Neutral', 'Greed', 'Extreme Greed']

pnl_by_sentiment = merged_filtered.groupby('classification')['Closed PnL'].agg(['mean','sum','count'])
pnl_by_sentiment = pnl_by_sentiment.reindex(order)
print(pnl_by_sentiment)

plt.figure(figsize=(10,5))
sns.boxplot(data=merged_filtered, x='classification', y='Closed PnL', order=order)
plt.title('Closed PnL Distribution by Market Sentiment')
plt.ylim(merged_filtered['Closed PnL'].quantile(0.05),
         merged_filtered['Closed PnL'].quantile(0.95))  # removes extreme outliers
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
merged_filtered['is_win'] = merged_filtered['Closed PnL'] > 0

win_rate = merged_filtered.groupby('classification')['is_win'].mean() * 100
win_rate = win_rate.reindex(order)

plt.figure(figsize=(8,4))
win_rate.plot(kind='bar', color='steelblue')
plt.title('Win Rate % by Market Sentiment')
plt.ylabel('Win Rate %')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
print(win_rate)
```

```python
volume = merged_filtered.groupby('classification')['Size USD'].sum().reindex(order)

plt.figure(figsize=(8,4))
volume.plot(kind='bar', color='coral')
plt.title('Total Trading Volume (USD) by Sentiment')
plt.ylabel('Total Size USD')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
top_accounts = merged_filtered.groupby('Account')['Closed PnL'].sum()\
               .sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_accounts.plot(kind='bar', color='green')
plt.title('Top 10 Accounts by Total Closed PnL')
plt.ylabel('Total PnL')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
side_sentiment = merged_filtered.groupby(['classification', 'Side'])['Closed PnL']\
                 .mean().unstack().reindex(order)

side_sentiment.plot(kind='bar', figsize=(10,5))
plt.title('Avg PnL: BUY vs SELL by Sentiment')
plt.ylabel('Avg Closed PnL')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
top_coins = merged_filtered['Coin'].value_counts().head(10)

plt.figure(figsize=(10,5))
top_coins.plot(kind='bar', color='purple')
plt.title('Top 10 Most Traded Coins')
plt.ylabel('Trade Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
print("=" * 50)
print(" KEY FINDINGS SUMMARY")
print("=" * 50)

print("\n1️.Avg Closed PnL per Sentiment:")
print(merged_filtered.groupby('classification')['Closed PnL'].mean().reindex(order).round(2))

print("\n2️.Win Rate per Sentiment (%):")
print((merged_filtered.groupby('classification')['is_win'].mean()*100).reindex(order).round(2))

print("\n3️.Total Trades per Sentiment:")
print(merged_filtered.groupby('classification')['Closed PnL'].count().reindex(order))

print("\n4️.Top 5 Most Traded Coins:")
print(merged_filtered['Coin'].value_counts().head(5))

print("\n5️.Unique Traders:", merged_filtered['Account'].nunique())
print("6️.Date Range:", merged_filtered['date'].min(), "to", merged_filtered['date'].max())
```

```python
# 'value' column is the numeric fear/greed score (0-100)
correlation = merged_filtered[['Closed PnL', 'value', 'Size USD', 'Fee']].corr()

plt.figure(figsize=(8,6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
#sentiment score vs PnL
print("Correlation between Sentiment Score & Closed PnL:",
      merged_filtered['value'].corr(merged_filtered['Closed PnL']).round(4))
```

```python
# Monthly PnL trend
merged_filtered['month'] = merged_filtered['date'].dt.to_period('M')

monthly_pnl = merged_filtered.groupby('month')['Closed PnL'].sum()

plt.figure(figsize=(14,5))
monthly_pnl.plot(kind='line', marker='o', color='blue')
plt.title('Total Closed PnL Over Time (Monthly)')
plt.ylabel('Total PnL')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sentiment_counts = merged_filtered.groupby(['month','classification']).size().unstack(fill_value=0)

sentiment_counts.plot(kind='bar', stacked=True, figsize=(14,5))
plt.title('Market Sentiment Distribution Over Time')
plt.ylabel('Trade Count')
plt.xlabel('Month')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Best & Worst Performing Coins by Sentiment
coin_sentiment = merged_filtered.groupby(['Coin','classification'])['Closed PnL']\
                 .mean().unstack().reindex(columns=order)

# Top 10 coins by total PnL
top_coins_pnl = merged_filtered.groupby('Coin')['Closed PnL'].sum()\
                .sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
top_coins_pnl.plot(kind='bar', color='teal')
plt.title('Top 10 Coins by Total Closed PnL')
plt.ylabel('Total PnL')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bottom 10 coins
worst_coins_pnl = merged_filtered.groupby('Coin')['Closed PnL'].sum()\
                  .sort_values().head(10)

plt.figure(figsize=(10,5))
worst_coins_pnl.plot(kind='bar', color='red')
plt.title('Bottom 10 Coins by Total Closed PnL')
plt.ylabel('Total PnL')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
